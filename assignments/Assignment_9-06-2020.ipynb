### Question 1:
### Print ManOfMach count of each player in Hyderabad stadium


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/kotagiri-preeti/Summer-Internship-AI-and-ML/master/Match%20(1).csv",index_col="Match_SK")
df
```

```python
df.loc[df['City_Name' ]=='Hyderabad']['ManOfMach'].value_counts()
```

## Question 2:
### Plot a pie chart for the country_name column


```python
df.groupby(['Country_Name']).sum().plot(kind = 'pie' , y='Season_Year', startangle=45,radius=1.5, autopct='%1.1f%%')
plt.show()
```
